***9.drop irrelvent columns from the dataset using drop() functions***.

```python
car=car.drop(['Year','Seller_Type'],axis=1)
car.head(5)
```

***10.use renames() function to rename the column***

***11.print the total number of duplicate rows***

```python
duplicate_rows_car=car[car.duplicated()]
print("Number of duplicate rows:",duplicate_rows_car.shape)
```

```python
car.count()
```

***12.remove the duplicate rows using the drop_duplicates() function.***

```python
car=car.drop_duplicates()
car.head
```

***13.drop the missing values from the dataset***

```python
print(car.isnull().sum())
```

```python
#drop missing values
car=car.dropna()
car.count()
```

***14.plot a histogram to find the number of cars per brand***

```python
car.Price.value_counts().nlargest(40).plot(kind='bar',figsize=(10,5))
plt.title("Numbers of cars by price ")
plt.ylabel('number of cars')
plt.xlabel("price");
```

***15.draw a correlation plot between the variables***

```python
plt.figure(figsize=(20,10))
c=car.corr()
sns.heatmap(c,cmap="BrBG",annot=True)
c
```
